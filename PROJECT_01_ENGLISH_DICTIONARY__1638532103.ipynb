```python
import json
items = 1

while(items !=0):
    print('*******************Welcome to English Dictionary****************************')
    items=int(input(''' Enter                1: Add a new word
                      2: Find the meaning
                      3: Update a word
                      4: Exit

                       enter your choice: '''))
    if(items==1):
        temp1={}
        temp2={}
        word=input("enter new word  :").title()
        with open('words.txt','r')as file:
            x=json.load(file)
            for j in x:
                if(word==j):
                    print('word in dictionary :',word,'is :',x[j])
                    break
            else:
                meaning=input('Enter the meaning: ').title()
                temp1[word]=meaning
        with open('words.txt','r')as g:
            a=json.load(g)
            for i in a,temp1:
                temp2.update(i)
        with open('words.txt','w')as f:
            json.dump(temp2,f)
                        
                
    elif(items==2):
        word=input("find the meaning :").title()
        with open('words.txt','r')as file:
            x=json.load(file)
            for j in x:
                if (word==j):
                    print('the meaning of :',word,'is :',x[j])
                    break
            else:
                print('wrong word or enter correct word')
                
                
    elif(items==3):
        temp1={}
        temp2={}
        word=input('enter word to update :').title()
        with open('words.txt','r')as file:
            x=json.load(file)
            if word in x:
                meaning=input('Enter the meaning: ')
                temp1.update({word:meaning})
            else:
                print("Word is not in dictionary try adding")
                break
        for i in x,temp1:
            temp2.update(i)
        with open('words.txt','w')as f:
            json.dump(temp2,f)
                        
           
                    
    elif(items==4):
        print('Exit')
        break
        
    else:
        print('enter the correct choice 1 to 4')
        print('********************************Thank you***********************')
```
